```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
options = Options()
options.headless = False
options.add_experimental_option('detach', True)
```

```python
browser = webdriver.Chrome(ChromeDriverManager().install(), options = options)

browser.maximize_window()
browser.get(url)
browser.set_page_load_timeout(10)
```

```python
reviews_html = browser.page_source
reviews_html = bs(reviews_html, 'lxml') #allows further parsing 
reviews_html
```

```python
reviews = reviews_html.find_all('div', class_='review')
for review in reviews:
    usr = review.find('span', class_='a-profile-name').text
    stars = review.find('span', class_='a-icon-alt').text
    text = review.find('span', class_='review-text-content').text
    likes = review.find('span', class_='cr-vote-text').text
    print(likes)
    break
```
